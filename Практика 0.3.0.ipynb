# **Задача 5.**

Создайте игру "Палач" или "Виселица", в которой пользователь должен угадать слово, предлагая буквы. Программа должна выбрать случайное слово, а затем разрешить пользователю угадывать буквы. После каждого угадывания программа должна сообщать пользователю, есть ли в слове буква или нет.

**Этап 1**
Создать саму игру (игра должна иметь привлекательный интерфейс, соотвесвующий классической висилице)

**Этап 2:**
Ввести подсчет количества очков и ограничения на попытки

**Этап 3:** Добавить возможность сразу угадать слово

````python
import random
words = ['волна', 'жрица', 'петух', 'птица','маржа','совок','фирма','гриль','лавка','взлом','удача','борец','тотем','благо','створ','кусок','занос']
rword = random.choice(words)
word_correct = "_" * len(rword)
error = []
correct = []
badcount = 0
point = 0
while badcount < 7:
  char = input('Буква или слово целиком - ')
  if char in rword:
    if char == rword:
      answered = len([char for char in rword if char in correct])
      print('Вы отгадали слово целиком:', rword, '\n Ваше кол-во очков', point + (len(rword) - answered) * 10)
      break
    correct.append(char)
    word_correct = " ".join([char if char in correct else "_" for char in rword])
    print('Слово:',word_correct, '\n Ошибки:', *error,  '\n Попадания:', *correct)
    point += (10 - badcount) * rword.count(char)
  else:
    badcount += 1
    error.append(char)
    print('Слово:', word_correct,'\n Ошибки:', *error, '\n Попадания:', *correct)
  print()
  if set(word_correct) == set(rword + " "):
    print('Вы отгадали слово:', rword, '\n Ваше кол-во очков', point)
    break
if badcount >= 9:
  print()
  print('Вы не отгдали слово', rword)

````
